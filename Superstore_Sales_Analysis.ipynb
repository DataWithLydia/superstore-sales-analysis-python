```python
import matplotlib.pyplot as plt
```

```python
sales_by_category = df.groupby('Category')['Sales'].sum()

sales_by_category.plot(kind='bar')

plt.title('Sales by Category')
plt.xlabel('Category')
plt.ylabel('Sales')
plt.show()
```

```python
plt.savefig('sales_by_category.png')
```

```python
## Insights

- Technology generated the highest sales among all categories.
- Office Supplies generated the lowest sales.
- Technology contributes significantly to overall revenue.
- Further analysis can determine whether high sales also translate to high profit.
```

```python
profit_by_category = df.groupby('Category')['Profit'].sum()

profit_by_category.plot(kind='bar')

plt.title('Profit by Category')
plt.xlabel('Category')
plt.ylabel('Profit')
plt.show()
```

```python
plt.savefig('profit_by_category.png')
```

```python
## Insights

- Technology generated the highest profit among all categories.
- Office Supplies was the second most profitable category.
- Furniture generated the lowest profit despite having high sales.
- High sales do not always translate into high profit.
```

```python
sales_by_region = df.groupby('Region')['Sales'].sum()

sales_by_region.plot(kind='bar')

plt.title('Sales by Region')
plt.xlabel('Region')
plt.ylabel('Sales')

plt.show()
```

```python
plt.savefig('sales_by_region.png')
```

```python
## Insights

- West region recorded the highest sales.
- East region was the second-best performing region.
- South region generated the lowest sales.
- Sales performance varies significantly across regions.
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
```

```python
monthly_sales = df.groupby(df['Order Date'].dt.month)['Sales'].sum()

monthly_sales.plot(kind='line', marker='o')

plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Sales')

plt.show()
```

```python
## Insights

- Sales increased towards the end of the year.
- November recorded the highest sales.
- February recorded the lowest sales.
- The business shows strong seasonality, with peak performance in Q4.
```
